```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
pwd = os.getcwd()

os.chdir(pwd+'/thin_zp_rotate/')
f_1      = []
angles_1 = []
maxvals_1 = []
for i in np.arange(1,51):
    angle = 0.1*i
    temp_ = np.load('foc_spot_1_micron_'+str(round(angle,2))+'_degree.npy',)
    angles_1.append(angle)
    f_1.append(temp_)
    maxvals_1.append(np.max(np.abs(temp_)))
    
os.chdir(pwd+'/thick_zp_rotate/')
f_2      = []
angles_2 = []
maxvals_2 = []
for i in np.arange(1,41):
    angle = 0.1*i
    temp_ = np.load('foc_spot_Q_3.33_'+str(round(angle,2))+'_degree.npy',)
    angles_2.append(angle)
    f_2.append(temp_)
    maxvals_2.append(np.max(np.abs(temp_)))

os.chdir(pwd)
```

```python
plt.imshow(f_1[1])
plt.colorbar()
print(np.max(np.abs(f_1))**2)
```

```python
norm_ = 475
data_1 = np.abs(np.array(maxvals_1))
#data_1 = data_1/np.max(data_1)
data_1 = data_1/norm_
data_2 = np.abs(np.array(maxvals_2))
#data_2 = data_2/np.max(data_2)
data_2 = data_2/norm_

_norm_ = 225000
_data_1 = np.abs(np.array(maxvals_1))**2
#_data_1 = _data_1/np.max(_data_1)
_data_1 = _data_1/_norm_
_data_2 = np.abs(np.array(maxvals_2))**2
#_data_2 = _data_2/np.max(_data_2)
_data_2 = _data_2/_norm_


plt.rcParams["figure.figsize"] = (12,6)
fig,(ax1,ax2) = plt.subplots(1,2)
ax1.plot(angles_1,data_1,'g*-',label="Q=0.33")
ax1.plot(angles_2,data_2,'m^-',label="Q=3.33")
ax1.set_title("Maximum Value in focal plane",size=12)
ax1.set_ylabel("Normalized units",size=12)
ax1.set_xlabel("Tilt angle in degrees",size=12)
ax1.set_ylim((-0.1,1.1))
ax1.legend()
ax2.plot(angles_1,_data_1,'g*-',label="Q=0.33")
ax2.plot(angles_2,_data_2,'m^-',label="Q=3.33")
ax2.set_title("Maximum Intensity in focal plane",size=12)
ax2.set_ylabel("Normalized units",size=12)
ax2.set_xlabel("Tilt angle in degrees",size=12)
ax2.set_ylim((-0.1,1.1))
ax2.legend()
plt.savefig("final_tilt_plot.svg",format='svg',dpi=250)
plt.show()
```
