```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
f       = []
angles  = []
diff    = []
sums    = []
maxvals = []
os.chdir(os.getcwd()+'/results')
for i in np.arange(0,50):
    angle = 0.1*i
    temp_ = np.load('foc_spot_Q_3.33_'+str(round(angle,2))+'_degree.npy',)
    temp_ = np.abs(temp_)**2
    angles.append(angle)
    f.append(temp_)
    maxvals.append(np.max(temp_))
    sums.append(np.sum(temp_))
    diff.append(np.sum(temp_-f[0]))
```

```python
I1 = f[0][250-25:250+25,250-25:250+25]
I2 = f[-1][250-25:250+25,250-25:250+25]
plt.rcParams["figure.figsize"] = (12,10)
fig,ax = plt.subplots(2,2)
im1 = ax[0,0].imshow(I1,cmap='jet')
plt.colorbar(im1,ax=ax[0,0])
im2 = ax[0,1].imshow(I2,cmap='jet')
plt.colorbar(im2,ax=ax[0,1])
a_ = int(np.shape(I1)[0]/2)
ax[1,0].plot(I1[:,a_],'b--')
ax[1,0].plot(I1[a_,:],'g--')
ax[1,1].plot(I2[:,a_],'b--')
ax[1,1].plot(I2[a_,:],'g--')
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (12,6)
fig,(ax1,ax2) = plt.subplots(1,2)
ax1.plot(I1[:,a_],'b--')
ax1.plot(I2[:,a_],'g--')
ax2.plot(I1[a_,:],'b--')
ax2.plot(I2[a_,:],'g--')
print(np.max(I1))
print(np.max(I2))
```

```python
plt.rcParams["figure.figsize"] = (12,6)
fig,(ax1,ax2) = plt.subplots(1,2)
ax1.plot(angles,sums/np.max(sums),'*-')
ax1.set_title("Sum of [200x200] pixels around focal spot")
ax1.set_ylabel("Normalized units")
ax1.set_xlabel("Tilt angle in degrees")
ax2.plot(angles,maxvals/np.max(maxvals),'*-')
#ax2.plot(angles,maxvals,'*-')
ax2.set_title("Maximum value in focal plane")
ax2.set_ylabel("Normalized units")
ax2.set_xlabel("Tilt angle in degrees")
plt.show()
```
